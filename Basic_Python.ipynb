# 11. Given a two list of numbers (hardcoded in the program), create a new list such that new list should contain only odd numbers from the first list and even numbers from the second list.

```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]
list3 = []

#this for loop used for list1 
for x in list1:
  #check the number is odd or even. if odd then condition becomes true and put that value in the new list
  if(x%2!=0):
    list3.append(x)

#this for loop used for list2
for q in list2:
  #check the number is odd or even. if even then condition becomes true and put that value in the new list
  if(q%2==0):
    list3.append(q)

print(list3)
```

Given a string, write a python program to check if it is palindrome or not. Define a function named palindrome_checker_<your-student-id> in your program.

```python
def palindrome_checker(i,st,n):

  #when the i reach the middle of the string it will return true 
  if i>=n/2:
    return True
  
  #if both string are not match then it will return false that means not palindrome
  elif st[i] != st[n-i-1]:
    return False
  
  #this function call itself 
  return palindrome_checker(i+1,st,n)

#user enter a string 
s = str(input())

#gets length of the string
p = len(s)

#the string becomes lower case
s = s.lower()

#in if condition palindrome_checker function is call
if palindrome_checker(0,s,p) == True:
  print("It is a Palindrome")
else:
  print("Not a Palindrome")
```
